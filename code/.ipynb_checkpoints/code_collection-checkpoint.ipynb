```python
import ast
import requests
import json
import pandas as pd
import re
from urllib.request import Request, urlopen
import urllib.request
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
import time
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome("C:/Users/USER/Desktop/chromedriver_win32/chromedriver.exe",chrome_options=chrome_options) 
```

```python
driver.get("https://www.persee.fr/collections")
```

```python
collection = pd.DataFrame(index=list(np.arange(326)), columns=['NomCollection','Nb_Numero','Nb_Doc','Annee_creation','Annee_fin'])

list_nc = []
list_nbd = []
list_nbn = []
list_ac = []
list_f= []


for i in range(326):
    nomCollection = driver.find_elements_by_class_name("collection-title")[i].text
    details =  driver.find_elements_by_class_name("collection-details")[i].text
    
    try :
        nb_numero = int(((re.search("[0-9]+"+" numéros",str(details))).group(0))[:-8])
    except:
        nb_numero=None
        pass
    
    
    try :
        nb_doc = int((((re.search("[0-9]+"+ " documents",str(details)))).group(0))[:-10])
    except:
        nb_doc=None
        pass
    
    try :
    	annee_crea = (((re.search("documents, "+"[0-9]+",str(details))).group(0))[11:])
    except:
        annee_crea = None
        pass
    
    try :
        annee_fin = (((re.search("-"+"[0-9]+",str(details))).group(0))[1:])
    except:
        annee_fin = None
        pass

    list_nc.append(nomCollection)
    list_nbn.append(nb_numero)
    list_ac.append(annee_crea)
    list_f.append(annee_fin)
    list_nbd.append(nb_doc)

collection["NomCollection"] = list_nc
collection["Nb_Numero"] = list_nbn
collection["Nb_Doc"] = list_nbd
collection["Annee_creation"] =list_ac
collection["Annee_fin"] =list_f
```

```python
collection = collection.replace('Hommes & Migrations', 'Hommes et Migrations')
```
